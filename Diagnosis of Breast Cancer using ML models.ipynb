```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
```

```python
df
```

```python
df.columns
```

```python
mean_features = list(df.columns[1:11])
se_features = list(df.columns[11:21])
worst_features=list(df.columns[21:31])
```

```python
mean_features.append('diagnosis')
se_features.append('diagnosis')
worst_features.append('diagnosis')
```

```python
corr=df[mean_features].corr()
```

```python
corr
```

```python
prediction_vars=['radius_mean','perimeter_mean','area_mean','compactness_mean','concavity_mean','concave points_mean','radius_se','area_se','radius_worst','compactness_worst']
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.15,random_state=1)
```

```python
train_x=train[prediction_vars]
train_y=train['diagnosis']
test_x=test[prediction_vars]
test_y=test['diagnosis']
```

```python
train_x
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model1 = RandomForestClassifier()
model2 = KNeighborsClassifier()
model3 = SVC()
#model4 = MLPClassifier
```

```python
#Parameter tuning for RandomForest Classifier 
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_depth':(1,2,3,4), 'n_estimators':(10,50,100,500)}
best_model=GridSearchCV(model1,parameters)
```

```python
best_model.fit(train_x,train_y)
```

```python
best_model.best_params_
```

```python
model1.fit(train_x,train_y)
```

```python
prediction=model1.predict(test_x)
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y,prediction)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
print(precision_score(test_y, prediction))
print(recall_score(test_y,prediction))
print(accuracy_score(test_y,prediction))
```

```python
#Voting Classifier
from sklearn.ensemble import VotingClassifier
#create a dictionary of our models
estimators=[('RF', model1), ('KNN', model2), ('SVC', model3)]
#create our voting classifier, inputting our models
ensemble = VotingClassifier(estimators, voting='hard') 
```

```python
#fit model to training data
ensemble.fit(train_x, train_y)
```

```python
predict_ensemble = ensemble.predict(test_x)
```

```python
predict_ensemble
```

```python
confusion_matrix(test_y,predict_ensemble)
```

```python
print(precision_score(test_y, predict_ensemble))
print(recall_score(test_y,predict_ensemble))
print(accuracy_score(test_y,predict_ensemble))
```

```python
from sklearn.neural_network import MLPClassifier

# Create model object
clf = MLPClassifier(hidden_layer_sizes=(6,5),
                    random_state=5,
                    verbose=True,
                    learning_rate_init=0.01)

# Fit data onto the model
clf.fit(train_x,train_y)
```
